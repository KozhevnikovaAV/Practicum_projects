**Задание 5. В двух столбцах есть пропущенные значения. Один из них — `days_employed`. Пропуски в этом столбце вы обработаете на следующем этапе. Другой столбец с пропущенными значениями — `total_income` — хранит данные о доходах. На сумму дохода сильнее всего влияет тип занятости, поэтому заполнить пропуски в этом столбце нужно медианным значением по каждому типу из столбца `income_type`. Например, у человека с типом занятости `сотрудник` пропуск в столбце `total_income` должен быть заполнен медианным доходом среди всех записей с тем же типом.**

```python
for t in data['income_type'].unique():
    data.loc[(data['income_type'] == t) & (data['total_income'].isna()), 'total_income'] = \
    data.loc[(data['income_type'] == t), 'total_income'].median()
```

### Категоризация данных

**Задание 16. На основании диапазонов, указанных ниже, создайте в датафрейме `data` столбец `total_income_category` с категориями:**

- 0–30000 — `'E'`;
- 30001–50000 — `'D'`;
- 50001–200000 — `'C'`;
- 200001–1000000 — `'B'`;
- 1000001 и выше — `'A'`.


**Например, кредитополучателю с доходом 25000 нужно назначить категорию `'E'`, а клиенту, получающему 235000, — `'B'`. Используйте собственную функцию с именем `categorize_income()` и метод `apply()`.**

```python
def categorize_income(income):
    try:
        if 0 <= income <= 30000:
            return 'E'
        elif 30001 <= income <= 50000:
            return 'D'
        elif 50001 <= income <= 200000:
            return 'С'
        elif 200001 <= income <= 1000000:
            return 'B'
        elif income >= 1000001:
            return 'A'
    except:
        pass
```

```python
data['total_income_category'] = data['total_income'].apply(categorize_income)
```

**Задание 18. Создайте функцию, которая на основании данных из столбца `purpose` сформирует новый столбец `purpose_category`, в который войдут следующие категории:**

- `'операции с автомобилем'`,
- `'операции с недвижимостью'`,
- `'проведение свадьбы'`,
- `'получение образования'`.

**Например, если в столбце `purpose` находится подстрока `'на покупку автомобиля'`, то в столбце `purpose_category` должна появиться строка `'операции с автомобилем'`.**

**Используйте собственную функцию с именем `categorize_purpose()` и метод `apply()`. Изучите данные в столбце `purpose` и определите, какие подстроки помогут вам правильно определить категорию.**

```python
def categorize_purpose(row):
    try:
        if 'автом' in row:
            return 'операции с автомобилем'
        elif 'жил' in row or 'недвиж' in row:
            return 'операции с недвижимостью'
        elif 'свад' in row:
            return 'проведение свадьбы'
        elif 'образов' in row:
            return 'получение образования'
    except:
        return 'нет категории'
```

```python
data['purpose_category'] = data['purpose'].apply(categorize_purpose)
```

### Шаг 3. Исследуйте данные и ответьте на вопросы

#### 3.1 Есть ли зависимость между количеством детей и возвратом кредита в срок?

```python
def child_category (children):
    if children == 0:
        return 'Нет детей'
    if children == 1:
        return 'Один ребенок'
    if children >= 2:
        return 'Двое или более детей'
    
data['children'] = data['children'].apply(child_category)

children_credit = data.pivot_table(index='children', values='debt', aggfunc=['count', 'sum', 'mean'])
children_credit.columns = ['Количество клиентов', 'Количество просрочек', 'Доля просрочек']
children_credit
```

**Вывод:** Ввиду проведенного исследования, можно сделать вывод о том, что при появлении ребенка в семье доля просрочек, в отличие от клиентов без детей, растет, и с появлением 2-го или больше детей в семье так же возрастает.

#### 3.2 Есть ли зависимость между семейным положением и возвратом кредита в срок?

```python
family_status_credit = data.pivot_table(index='family_status', values='debt', aggfunc=['count', 'sum', 'mean'])
family_status_credit.columns = ['Количество клиентов', 'Количество просрочек', 'Доля просрочек']
family_status_credit.sort_values(by='Доля просрочек', ascending=False)
```

**Вывод:** Ввиду проведенного исследования, можно сделать вывод о том, что свободные клиенты (никогда не состоящие в зарегистрированном браке) менее ответственные, так как у них самая высокая доля просрочек по кредитам (0.097639), в то время как у других клиентов (состоящих в зарегистрированном браке на данный момент или же состоящих в зарегистрированном браке ранее) доля просрочек по кредитов составляет от 0.075606 и стремится вниз.

#### 3.3 Есть ли зависимость между уровнем дохода и возвратом кредита в срок?

```python
def categorize_income(total_income):
    if 0 <= total_income <= 100000:
        return 'доход до 100000'
    if 100001 <= total_income <= 300000:
        return 'доход до 300000'
    if 300001 <= total_income <= 500000:
        return 'доход до 500000'
    if 500001 <= total_income <= 700000:
        return 'доход до 700000'
    if 700001 <= total_income <= 900000:
        return 'доход до 900000'
    if total_income >= 1000001:
        return 'доход от 1000000'
        
data['total_income'] = data['total_income'].apply(categorize_income)

total_income_credit = data.pivot_table(index='total_income', values='debt', aggfunc=['count', 'sum', 'mean'])
total_income_credit.columns = ['Количество клиентов', 'Количество просрочек', 'Доля просрочек']
total_income_credit.sort_values(by='Доля просрочек', ascending=False)
```

**Вывод:** Ввиду проведенного исследования, можно сделать вывод о том, что зависимости между уровнем дохода и возвратом кредита в срок нет. Исследование проводилась по категориями клиентов в зависимости от их дохода, однако хочу отметить, что доля просрочек у клиентов с большими доходами больше, чем у клиентов с меньшим.

#### 3.4 Как разные цели кредита влияют на его возврат в срок?

```python
def categorize_purpose(row):
    try:
        if 'автом' in row:
            return 'операции с автомобилем'
        elif 'жил' in row or 'недвиж' in row:
            return 'операции с недвижимостью'
        elif 'свад' in row:
            return 'проведение свадьбы'
        elif 'образов' in row:
            return 'получение образования'
    except:
        return 'нет категории'
    
data['purpose_category'] = data['purpose'].apply(categorize_purpose)
data['purpose_category']

purpose_category_credit = data.pivot_table(index='purpose_category', values='debt', aggfunc=['count', 'sum', 'mean'])
purpose_category_credit.columns = ['Количество клиентов', 'Количество просрочек', 'Доля просрочек']
purpose_category_credit
```
